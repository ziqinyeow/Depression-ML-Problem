```python
depression_question_no = [3, 5, 10, 13, 16, 17, 21, 24, 26, 31, 34, 37, 38, 42]
depression_col = []

for n in depression_question_no:
    depression_col.append(f"Q{n}A")

df = survey["basic"].filter(depression_col)
df = df.subtract(1, axis=1)
```

```python
df["depression_count"] = df[depression_col].sum(axis=1)
```

```python
df.head()
```

## Identify y-label

```python
def condition(depression_score):
    if depression_score <= 9:
        return 'normal'
    if 10 <= depression_score <= 13:
        return 'mild'
    if 14 <= depression_score <= 20:
        return 'moderate'
    if 21 <= depression_score <= 27:
        return 'severe'
    if depression_score > 28:
        return 'extremely severe'
```

```python
df["depression_condition"] = df["depression_count"].apply(condition)
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=df.sort_values('depression_condition')["depression_condition"])
plt.title('People Condition of Depression Level', fontsize=15)
plt.show()
```
